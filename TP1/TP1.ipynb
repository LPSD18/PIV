```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
from aux_func import bwLabel
from aux_func import psColor
```

```python
path = 'treino/'

lego = []
lego_files = [f for f in os.listdir(path) if f.endswith('.jpg')]

for file in lego_files:
    lego_path = path + file
    image = cv2.imread(lego_path)

    if image is not None:
        lego.append(image)
    else:
        print(f"Unable to read image: {file}")

plt.figure(figsize=(12,12))
x = 1
for image in lego:
    plt.subplot(4,4,x)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    x+=1
```

```python
grayLegos = []

for color in lego:
    grayImage = cv2.cvtColor(color,cv2.COLOR_RGB2GRAY)
    grayLegos.append(grayImage)

plt.figure(figsize=(12,12))
x = 1
for image in grayLegos:
    plt.subplot(4,4,x)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    x+=1
```
